```python
import geopandas as gpd
from shapely import Polygon
from rasterio.crs import CRS
import rioxarray as rxr
from rasterio.warp import reproject
```

```python
area_polygon = Polygon([(12, 42.56), (27, 42.56), (27, 51), (12, 51)])
area = gpd.GeoDataFrame(geometry=[area_polygon], crs=4326)
```

```python
(27-12)/(51-42.56)
```

```python
path = "../data/dnb_land_ocean_ice.2012.13500x13500.C1_geo.tif"
```

```python
raster = rxr.open_rasterio(path, masked=True).squeeze()
```

```python
clipped_data = raster.rio.clip(area.geometry)
clipped_data.rio.crs.to_proj4()
```

```python
clipped_data.rio.to_raster("../figures/nasa.tif")
```
